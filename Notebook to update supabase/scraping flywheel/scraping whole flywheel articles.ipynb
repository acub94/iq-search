```python
from bs4 import BeautifulSoup
import requests
```

```python
html = requests.get('https://flywheeloutput.com/sitemap.xml')
```

```python
data = BeautifulSoup(html.text, "xml")
```

```python
flywheel_urls = data.find_all('loc')
```

```python
flywheel_urls
```

```python
flywheel_urls.pop(0)
flywheel_urls.pop(0)
```

```python
flywheel_urls
```

```python
urls = []
for url in flywheel_urls:
    urls.append(url.string)
```

```python
urls
```

```python
import pandas as pd 

df = pd.DataFrame(columns=['content','url'])

for url in urls:
    response = requests.get(url)
    if response.status_code == 200:
        data = BeautifulSoup(response.text,'html.parser')
        paras = data.find_all('p')
        paras = [para.string for para in paras if para.string is not None and para.string[:4] != 'http']
        content = " ".join(paras)
        row = pd.DataFrame({'content':[content],'url':[url]})
        df = df.append(row,ignore_index=True)
        
```
